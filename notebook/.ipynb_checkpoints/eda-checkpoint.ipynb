```python
#understand the distribution of target variable
sns.countplot(x="default payment next month", data=df)

plt.title("Default Distribution")

plt.show()
```

Note:
The graph shows the distribution of customers who defaulted and did not default on their credit card payment next month. Most customers did not default (0), while a smaller number defaulted (1). This indicates that the dataset is imbalanced, with more non-default cases than default cases.

```python
#Age Distribution
plt.hist(df["AGE"], bins=20)

plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")

plt.show()
```

Note:
The age distribution shows that most customers in the dataset are between 25 and 40 years old, with the highest concentration around the early 30s. The number of customers gradually decreases as age increases, and there are fewer customers above 50 years old. This indicates that the dataset mainly consists of young to middle-aged credit card users.

```python
#credit limit distribution
plt.hist(df["LIMIT_BAL"], bins=30)

plt.title("Credit Limit Distribution")
plt.xlabel("Credit Limit")
plt.ylabel("Count")

plt.show()
```

Note:
The credit limit distribution shows that most customers have lower to moderate credit limits. As the credit limit increases, the number of customers decreases. Only a small number of customers have very high credit limits. This indicates that most credit card users in the dataset are given relatively moderate credit limits, while high credit limits are less common.

```python
#credit limit vs default
sns.boxplot(x="default payment next month", y="LIMIT_BAL", data=df)

plt.title("Credit Limit vs Default")

plt.show()
```

Note: 
The box plot compares the credit limit of customers who defaulted and those who did not. Customers who did not default (0) generally tend to have slightly higher credit limits, while customers who defaulted (1) tend to have lower credit limits. However, there is still overlap between the two groups, indicating that credit limit alone may not fully determine whether a customer will default.

```python
education_col = {
0: "Unknown",
1: "Graduate School",
2: "University",
3: "High School",
4: "Others",
5: "Unknown",
6: "Unknown"
}

df["EDUCATION_LABEL"] = df["EDUCATION"].map(education_col)
```

```python
#Education Distribution
sns.countplot(x="EDUCATION_LABEL", data=df)
plt.xticks(rotation=45)
plt.show()
```

Note:

The education level distribution shows that most customers belong to university (2) and graduate school (1) categories. A smaller number of customers have high school education (3), while very few customers fall into the others (4) category. According to the dataset description, the education variable is coded as 1 = graduate school, 2 = university, 3 = high school, and 4 = others. This indicates that the majority of credit card users in the dataset have higher education levels.

```python
marriage_col = {
1: "Married",
2: "Single",
3: "Others"
}

df["MARRIAGE_LABEL"] = df["MARRIAGE"].map(marriage_col)
```

```python
sns.countplot(x="MARRIAGE_LABEL", data=df)

plt.title("Marriage Status Distribution")

plt.xlabel("Marriage Status")
plt.ylabel("Count")

plt.show()
```

Note:

The marriage status distribution shows that most customers in the dataset are single, followed by married customers. A small number of customers belong to the others category. According to the dataset description, the marriage variable is coded as 1 = married, 2 = single, and 3 = others.

### Note:
The repayment status variable PAY_0 represents the most recent repayment behavior of customers. It indicates how late a customer was in their latest payment. Since it reflects the most recent financial behavior, it can provide useful insight into whether a customer might default in the next month. Other repayment variables such as PAY_2 to PAY_6 represent repayment status in earlier months.

Later in the analysis or modeling stage, all repayment variables (PAY_0 to PAY_6) can be considered together to understand customers' payment history over time

```python
#check repayment status
sns.countplot(x="PAY_0", data=df)

plt.title("Repayment Status PAY_0")

plt.show()
```

Note:

The repayment status distribution for PAY_0 shows the recent payment behavior of customers. Most customers have a value of 0, which indicates that they paid their credit card bill on time. Some customers have values such as 1 or 2, meaning they were delayed by one or two months in their payments. A smaller number of customers show higher delay values, indicating longer repayment delays. This suggests that while most customers pay on time, a portion of customers experience payment delays which may increase the risk of default.
